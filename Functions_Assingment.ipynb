Q1) What is the difference between a function and a method in Python?
Answer: Function is independent block of code which defined by 'def' keyword,
function can use anywhere in the given program.
Method is denoted by 'self:' Methods take an implicit self parameter, which refers to the instance of the class, methods can access and modify the attributes of the class.

```python
# Function
def greet(name):
    print(f"Hello, {name}!")

# Class with a method
class Person:
    def __init__(self, name):
        self.name = name

    def greet(self):
        print(f"Hello, my name is {self.name}!")

# Call the function
greet("KARTIK")

# Call the method
person = Person("KIWI")
person.greet()
```

Q6) Explain the concept of generators in Python and how they are defined.
Answer: Generators are a type of iterable, similar to lists or tuples unlike lists, generator generate values as they're needed. By using "yeild" statement in function we can create generators.

```python
def my_generator():
    num = 0
    while True:
        yield num
        num += 1
gen = my_generator()
for _ in range(10):
    print(next(gen))
```

Q9) Explain the purpose and usage of the "map()" function in Python.
Answer: The "map()" function in Python is a pre-defined function that applies a given function to each item of an iterable, This allows you to perform operations on each item in the iterable without having to write a loop.

```python
def square(x):
    return x ** 2

numbers = [1, 2, 3, 4, 5]
squared_numbers = map(square, numbers)

print(list(squared_numbers))
```

Q10) What is the difference between "map()", "reduce()", and "filter()" functions in Python?
Answer: Map Function allows you to perform operations on each item in the iterable without having to write a loop. Reduce function is applied to the first two items in the iterable, then to the result and the next item, and so on, until the iterable is exhausted. Filter function is applied to each item in the iterable, and only those items for which the function returns True are included in the resulting filter object.

```python
from functools import reduce
numbers = [1, 2, 3, 4, 5]

# Map()
squared_numbers = list(map(lambda x: x ** 2, numbers))
print(squared_numbers)  # Output: [1, 4, 9, 16, 25]

# Reduce()
sum_of_numbers = reduce(lambda x, y: x + y, numbers)
print(sum_of_numbers)  # Output: 15

# Filter()
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(even_numbers)  # Output: [2, 4]
```

```python
#Q1)Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.
def sum_even_numbers(numbers):
    even_sum = 0
    for num in numbers:
        if num % 2 == 0:
            even_sum += num
    return even_sum

numbers = input("Enter a list of numbers separated by spaces: ")
numbers = [int(num) for num in numbers.split()]
result = sum_even_numbers(numbers)
print("The sum of even numbers is:", result)
```

```python
#Q4) Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(n):
    if n <= 1:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    max_divisor = int(n**0.5) + 1
    for d in range(3, max_divisor, 2):
        if n % d == 0:
            return False
    return True
for i in range(1, 201):
    if is_prime(i):
        print(i, "is a prime number")
```

```python
#Q5) Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
def fibonacci(n):
    a, b = 0, 1
    for i in range(n):
        print(a)
        a, b = b, a + b

# Example usage:
n = 10
print("Fibonacci series up to {} terms:".format(n))
fibonacci(n)
```

```python
#Q6) Write a generator function in Python that yields a powers of 2 up to a given exponent.
def powers_of_two(n):
    for i in range(n+1):
        yield 2 ** i

# Example usage:
for power in powers_of_two(5):
    print(power)
```

```python
#Q9) Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

celsius_temps = [0, 10, 20, 30, 40]

fahrenheit_temps = list(map(celsius_to_fahrenheit, celsius_temps))

print("Celsius Temperatures:", celsius_temps)
print("Fahrenheit Temperatures:", fahrenheit_temps)
```

```python
#Q10) Create a Python program that uses `filter()` to remove all the vowels from a given string.
def is_vowel(char):
    vowels = 'aeiouAEIOU'
    return char in vowels
string = "Hello World"

no_vowels = ''.join(filter(lambda char: not is_vowel(char), string))

print("Original String:", string)
print("String without Vowels:", no_vowels)
```
